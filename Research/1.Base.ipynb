## English Transcribe

```python
import whisper

# Load Whisper model (choose: tiny, base, small, medium, or large)
model = whisper.load_model("base")  # "base" is good balance of speed & accuracy

# Specify your audio file path
audio_file = "Dataset/Pankaj sir audio.wav"  # Change this to your actual file

# Transcribe the audio
print("Transcribing audio...")
result = model.transcribe(audio_file)

# Save transcription to a text file
output_file = "output/transcription.txt"
with open(output_file, "w", encoding="utf-8") as f:
    f.write(result["text"])

print("\n✅ Transcription complete!")
print("Transcribed Text:\n", result["text"])
print(f"\n📄 Transcription saved to: {output_file}")
```

## Hindi Transcribe

```python
import whisper

# Load a better Whisper model for Hindi (medium or large is recommended)
model = whisper.load_model("medium")  # Use "large" for best accuracy

# Specify your Hindi audio file path
audio_file = "Dataset/Pankaj sir audio.wav"  # Change this to your actual file

# Transcribe the audio with language hint
print("Transcribing Hindi audio...")
result = model.transcribe(audio_file, language="hi")  # "hi" for Hindi

# Save transcription to a text file
output_file = "output/hinditranscription.txt"
with open(output_file, "w", encoding="utf-8") as f:
    f.write(result["text"])

print("\n✅ Transcription complete!")
print("Transcribed Text:\n", result["text"])
print(f"\n📄 Transcription saved to: {output_file}")
```
